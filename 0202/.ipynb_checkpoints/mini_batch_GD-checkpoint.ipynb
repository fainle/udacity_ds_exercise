```python
import numpy as np
np.random.seed(42) # 设置随机种子


def MSEStep(X, y, W, b, learn_rate = 0.005):
    y_pred = np.matmul(X, W) + b
    error = y - y_pred
     
    # compute steps
    W_new = W + learn_rate * np.matmul(error, X)
    b_new = b + learn_rate * error.sum()
    
#     print(W_new)
#     print(b_new)
    
    return W_new, b_new

def MAEStep(X, y, W, b, learn_rate = 0.005):
    pass


def miniBatchGD(X, y, batch_size = 20, learn_rate = 0.005, num_iter = 25, mothed="MSEStep"):
    """
    对数据集 执行小批次随机梯度下降
    X 预测特征数据
    y 结果值
    batch_size 每次迭代将采样多少个数据点
    Learn_rate：学习率
    num_iter：使用的批次数
    mothed: 使用的算法
    """
    
    n_points = X.shape[0] # 总数量
    W = np.zeros(X.shape[1]) # 初始化斜率
    b = 0 # 初始化
    regression_coef = [np.hstack((W,b))]
    
    # 执行指定批次的计算
    for _ in range(num_iter):
        batch = np.random.choice(range(n_points), batch_size)  # 随机取20个
        X_batch = X[batch,:]
        y_batch = y[batch]
        
        W, b = MSEStep(X_batch, y_batch, W, b, learn_rate)
        regression_coef.append(np.hstack((W,b)))
        
    return regression_coef


def main():
    regression_coef = miniBatchGD(X, y)
    print(regression_coef)
    
main()
```
