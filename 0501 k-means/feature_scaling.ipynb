```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import preprocessing as p

%matplotlib inline

plt.rcParams['figure.figsize'] = (16, 9)

import help2 as h

data = h.simulate_data(200, 2, 4)

df = pd.DataFrame(data)
df.columns = ['h', 'w']
df['h'] = np.abs(df['h'] * 100)
df['w'] = df['w'] * np.random.normal(50, 10, 200)
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df_ss = p.StandardScaler().fit_transform(df)
df_ss = pd.DataFrame(df_ss)
df_ss.columns = ['h', 'w']
```

```python
df_ss.head(5)
```

```python
plt.scatter(df_ss['w'], df_ss['h'])
```

```python
df_mm = p.MinMaxScaler().fit_transform(df)
df_mm = pd.DataFrame(df_mm)
df_mm.columns = ['h', 'w']
df_mm.head(5)
```

```python
plt.scatter(df_mm['w'], df_mm['h'])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(4)
model = kmeans.fit(data)
labels = model.predict(data)
```

```python
plt.scatter(df['w'], df['h'], c=labels, cmap='Set1')
```

```python
plt.scatter(df_mm['w'], df_mm['h'], c=labels, cmap='Set1')
```
