```python
import numpy as np
import matplotlib.pyplot as plt
import help2 as h

from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from IPython import display

%matplotlib inline

plt.rcParams['figure.figsize'] = [16, 9]
```

```python
data = h.simulate_data(200, 5, 4)
```

```python
kmeans = KMeans(4)
model = kmeans.fit(data)
labels = model.predict(data)
```

```python
h.plot_data(data, labels)
```

```python
kmeans2 = KMeans(2)
model2 = kmeans2.fit(data)
labels2 = model2.predict(data)
h.plot_data(data, labels2)
```

```python
scores = []
for center in range(1, 11):
    kmeans = KMeans(center)
    
    model = kmeans.fit(data)
    score = model.score(data)
    scores.append(abs(score))
    
centers = list(range(1, 11))
plt.plot(centers, scores)
plt.title('Scores Plot')
plt.xlabel('Centers')
plt.ylabel('Average Distance From Centroid')
```
